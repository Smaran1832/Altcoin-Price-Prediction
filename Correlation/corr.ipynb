```python
import pandas as pd
```

## Bitcoin

```python
bitcoin = pd.read_csv('Bitcoin_OHLC.csv')
bitcoin = bitcoin.drop(columns=['Unnamed: 0'])
bitcoin['Bitcoin_Change'] = bitcoin['Close'] - bitcoin['Open']
bitcoin
```

```python
df = bitcoin.copy()
df.index = df['Date']
df = df.drop(columns=['Date'])
df = df.drop(columns=['Open', 'High', 'Low', 'Close'])
df
```
