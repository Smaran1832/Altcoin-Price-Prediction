```python
import pandas as pd
```

```python
LTC=pd.read_excel("../../../Litecoin/2022Litecoin_Original.xlsx")
LTC['Date'] = pd.to_datetime(LTC['Date'])
LTC.set_index('Date', inplace=True)
```

```python
LTC
```

```python
# List of columns to convert
columns_to_convert = ['Gold_Price', 'Oil_Price', 'Nasdaq_Close', 'S&P500_Close', 'BTC_Close','BTC_Volume','BTC_Market_Cap','Tweets_Count','Tweets_Vader_Sentiment','News_Vader_Sentiment']

# Applying the transformation
for column in columns_to_convert:
    LTC[column + '_binary'] = (LTC[column] > LTC[column].shift(1)).astype(int)
```

```python
LTC.drop(["Internet_Penetration_USA","Population_USA","GDP_USA","Inflation_USA","Nasdaq_Open","Nasdaq_High","Nasdaq_Low","Nasdaq_Adj Close","Nasdaq_Volume","S&P500_Open","S&P500_High","S&P500_Low","S&P500_Adj Close","BTC_Open","BTC_High","BTC_Low",'Gold_Price', 'Oil_Price', 'Nasdaq_Close', 'S&P500_Close', 'BTC_Close','S&P500_Volume',],inplace=True,axis=1)
```

```python
LTC
```

```python
columns_to_lag = ['Litecoin_Change'] #'Gold_Price_binary','Oil_Price_binary','Tweets_Count_binary','Tweets_Vader_Sentiment_binary','News_Vader_Sentiment_binary'

# Number of lags (days back) to create
number_of_lags = 5

# Create lag features for the selected columns
for column in columns_to_lag:
    for lag in range(1, number_of_lags + 1):
        LTC[f'{column}_lag_{lag}'] = LTC[column].shift(lag)

# Optionally, remove any rows with NaN values that were created due to shifting
LTC.dropna(inplace=True)
```

```python
LTC
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Selecting the columns to standardize
features = LTC[['BTC_Volume', 'BTC_Market_Cap', 'Tweets_Count', 'Tweets_Vader_Sentiment', 'News_Vader_Sentiment']]

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the features
features_scaled = scaler.fit_transform(features)

# Replace the original columns with the scaled columns
LTC[['BTC_Volume', 'BTC_Market_Cap', 'Tweets_Count', 'Tweets_Vader_Sentiment', 'News_Vader_Sentiment']] = features_scaled
```

```python
LTC
```

```python
LTC_market= pd.read_excel("../../Base_Datasets/Alt_Coin_Yearly_Rank.xlsx", sheet_name = 'Litecoin_Daily')
LTC_market['Date'] = pd.to_datetime(LTC_market['Date'])
LTC_market.set_index('Date', inplace=True)
```

```python
LTC_market
```

```python
# Convert columns to float, removing any non-numeric characters like commas
LTC_market['Open*'] = pd.to_numeric(LTC_market['Open*'].replace('[^\d.]', '', regex=True), errors='coerce')
LTC_market['High'] = pd.to_numeric(LTC_market['High'].replace('[^\d.]', '', regex=True), errors='coerce')
LTC_market['Low'] = pd.to_numeric(LTC_market['Low'].replace('[^\d.]', '', regex=True), errors='coerce')
LTC_market['Close**'] = pd.to_numeric(LTC_market['Close**'].replace('[^\d.]', '', regex=True), errors='coerce')
LTC_market['Volume'] = pd.to_numeric(LTC_market['Volume'].replace('[^\d.]', '', regex=True), errors='coerce')
LTC_market['Market Cap'] = pd.to_numeric(LTC_market['Market Cap'].replace('[^\d.]', '', regex=True), errors='coerce')
```

```python
LTC_market
```

```python
from ta import add_all_ta_features
from ta.utils import dropna

# Clean NaN values
LTC_market = dropna(LTC_market)

# Add all technical indicators
LTC_market = add_all_ta_features(
    LTC_market, open="Open*", high="High", low="Low", close="Close**", volume="Volume", fillna=True
)
```

```python
LTC_market
```

```python
from ta.momentum import RSIIndicator
from ta.trend import MACD
from ta.volatility import BollingerBands

# RSI
LTC_market['rsi'] = RSIIndicator(close=LTC_market['Close**'], window=14, fillna=True).rsi()

# MACD
macd = MACD(close=LTC_market['Close**'], window_slow=26, window_fast=12, window_sign=9, fillna=True)
LTC_market['macd'] = macd.macd()
LTC_market['macd_signal'] = macd.macd_signal()
LTC_market['macd_diff'] = macd.macd_diff()

# Bollinger Bands
indicator_bb = BollingerBands(close=LTC_market['Close**'], window=20, window_dev=2, fillna=True)
LTC_market['bb_bbm'] = indicator_bb.bollinger_mavg()
LTC_market['bb_bbh'] = indicator_bb.bollinger_hband()
LTC_market['bb_bbl'] = indicator_bb.bollinger_lband()
```

```python
LTC_market
```

```python
# Shift all columns except 'Date'
columns_to_shift = LTC_market.columns.difference(['Date'])
LTC_market[columns_to_shift] = LTC_market[columns_to_shift].shift(-1)

LTC_market.dropna(inplace=True)
```

```python
Litecoin_final_dataset=LTC_market.join(LTC,how='inner')
```
