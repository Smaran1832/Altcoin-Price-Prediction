```python
import pandas as pd
```

```python
XMR=pd.read_excel("../../../Monero/2022Monero_Original.xlsx")
XMR['Date'] = pd.to_datetime(XMR['Date'])
XMR.set_index('Date', inplace=True)
```

```python
XMR
```

```python
# List of columns to convert
columns_to_convert = ['Gold_Price', 'Oil_Price', 'Nasdaq_Close', 'S&P500_Close', 'BTC_Close','BTC_Volume','BTC_Market_Cap','Tweets_Count','Tweets_Vader_Sentiment','News_Vader_Sentiment']

# Applying the transformation
for column in columns_to_convert:
    XMR[column + '_binary'] = (XMR[column] > XMR[column].shift(1)).astype(int)
```

```python
XMR.drop(["Internet_Penetration_USA","Population_USA","GDP_USA","Inflation_USA","Nasdaq_Open","Nasdaq_High","Nasdaq_Low","Nasdaq_Adj Close","Nasdaq_Volume","S&P500_Open","S&P500_High","S&P500_Low","S&P500_Adj Close","BTC_Open","BTC_High","BTC_Low",'Gold_Price', 'Oil_Price', 'Nasdaq_Close', 'S&P500_Close', 'BTC_Close','S&P500_Volume',],inplace=True,axis=1)
```

```python
XMR
```

```python
columns_to_lag = ['Monero_Change'] #'Gold_Price_binary','Oil_Price_binary','Tweets_Count_binary','Tweets_Vader_Sentiment_binary','News_Vader_Sentiment_binary'

# Number of lags (days back) to create
number_of_lags = 5

# Create lag features for the selected columns
for column in columns_to_lag:
    for lag in range(1, number_of_lags + 1):
        XMR[f'{column}_lag_{lag}'] = XMR[column].shift(lag)

# Optionally, remove any rows with NaN values that were created due to shifting
XMR.dropna(inplace=True)
```

```python
XMR
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Selecting the columns to standardize
features = XMR[['BTC_Volume', 'BTC_Market_Cap', 'Tweets_Count', 'Tweets_Vader_Sentiment', 'News_Vader_Sentiment']]

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the features
features_scaled = scaler.fit_transform(features)

# Replace the original columns with the scaled columns
XMR[['BTC_Volume', 'BTC_Market_Cap', 'Tweets_Count', 'Tweets_Vader_Sentiment', 'News_Vader_Sentiment']] = features_scaled
```

```python
XMR
```

```python
XMR_market= pd.read_excel("../../Base_Datasets/Alt_Coin_Yearly_Rank.xlsx", sheet_name = 'Monero_Daily')
XMR_market['Date'] = pd.to_datetime(XMR_market['Date'])
XMR_market.set_index('Date', inplace=True)
```

```python
XMR_market
```

```python
# Convert columns to float, removing any non-numeric characters like commas
XMR_market['Open*'] = pd.to_numeric(XMR_market['Open*'].replace('[^\d.]', '', regex=True), errors='coerce')
XMR_market['High'] = pd.to_numeric(XMR_market['High'].replace('[^\d.]', '', regex=True), errors='coerce')
XMR_market['Low'] = pd.to_numeric(XMR_market['Low'].replace('[^\d.]', '', regex=True), errors='coerce')
XMR_market['Close**'] = pd.to_numeric(XMR_market['Close**'].replace('[^\d.]', '', regex=True), errors='coerce')
XMR_market['Volume'] = pd.to_numeric(XMR_market['Volume'].replace('[^\d.]', '', regex=True), errors='coerce')
XMR_market['Market Cap'] = pd.to_numeric(XMR_market['Market Cap'].replace('[^\d.]', '', regex=True), errors='coerce')
```

```python
XMR_market
```

```python
from ta import add_all_ta_features
from ta.utils import dropna

# Clean NaN values
XMR_market = dropna(XMR_market)

# Add all technical indicators
XMR_market = add_all_ta_features(
    XMR_market, open="Open*", high="High", low="Low", close="Close**", volume="Volume", fillna=True
)
```

```python
XMR_market
```

```python
from ta.momentum import RSIIndicator
from ta.trend import MACD
from ta.volatility import BollingerBands

# RSI
XMR_market['rsi'] = RSIIndicator(close=XMR_market['Close**'], window=14, fillna=True).rsi()

# MACD
macd = MACD(close=XMR_market['Close**'], window_slow=26, window_fast=12, window_sign=9, fillna=True)
XMR_market['macd'] = macd.macd()
XMR_market['macd_signal'] = macd.macd_signal()
XMR_market['macd_diff'] = macd.macd_diff()

# Bollinger Bands
indicator_bb = BollingerBands(close=XMR_market['Close**'], window=20, window_dev=2, fillna=True)
XMR_market['bb_bbm'] = indicator_bb.bollinger_mavg()
XMR_market['bb_bbh'] = indicator_bb.bollinger_hband()
XMR_market['bb_bbl'] = indicator_bb.bollinger_lband()
```

```python
XMR_market
```

```python
# Shift all columns except 'Date'
columns_to_shift = XMR_market.columns.difference(['Date'])
XMR_market[columns_to_shift] = XMR_market[columns_to_shift].shift(-1)

XMR_market.dropna(inplace=True)
```

```python
Monero_final_dataset=XMR_market.join(XMR,how='inner')
```
