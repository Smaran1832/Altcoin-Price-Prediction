```python
import pandas as pd
```

```python
XRP_market= pd.read_excel("../../Base_Datasets/Alt_Coin_Yearly_Rank.xlsx", sheet_name = 'XRP_Daily')
XRP_market['Date'] = pd.to_datetime(XRP_market['Date'])
XRP_market.set_index('Date', inplace=True)
```

```python
XRP_market
```

```python
# Convert columns to float, removing any non-numeric characters like commas
XRP_market['Open*'] = pd.to_numeric(XRP_market['Open*'].replace('[^\d.]', '', regex=True), errors='coerce')
XRP_market['High'] = pd.to_numeric(XRP_market['High'].replace('[^\d.]', '', regex=True), errors='coerce')
XRP_market['Low'] = pd.to_numeric(XRP_market['Low'].replace('[^\d.]', '', regex=True), errors='coerce')
XRP_market['Close**'] = pd.to_numeric(XRP_market['Close**'].replace('[^\d.]', '', regex=True), errors='coerce')
XRP_market['Volume'] = pd.to_numeric(XRP_market['Volume'].replace('[^\d.]', '', regex=True), errors='coerce')
XRP_market['Market Cap'] = pd.to_numeric(XRP_market['Market Cap'].replace('[^\d.]', '', regex=True), errors='coerce')
```

```python
from ta import add_all_ta_features
from ta.utils import dropna

# Clean NaN values
XRP_market = dropna(XRP_market)

# Add all technical indicators
XRP_market = add_all_ta_features(
    XRP_market, open="Open*", high="High", low="Low", close="Close**", volume="Volume", fillna=True
)
```

```python
XRP_market
```

# Future Close
Continued After Fragment

```python
XRP_market['Future_Close'] = XRP_market['Close**'].shift(-1)  # Shift closing price to the previous day for comparison
XRP_market['Price_Direction'] = (XRP_market['Future_Close'] > XRP_market['Close**']).astype(int)  # 0 if price increases, else 1 
XRP_market.dropna(inplace=True)
```

```python
XRP_market.drop(["Future_Close"],axis=1,inplace=True)
```

```python
XRP_market
```

```python
# Shift all columns except 'Date'
columns_to_shift = XRP_market.columns.difference(['Date','Price_Direction'])
XRP_market[columns_to_shift] = XRP_market[columns_to_shift].shift(-1)

XRP_market.dropna(inplace=True)
```
