```python
import pandas as pd
```

```python
XLM=pd.read_excel("../../../Stellar/2022Stellar_Original.xlsx")
XLM['Date'] = pd.to_datetime(XLM['Date'])
XLM.set_index('Date', inplace=True)
```

```python
XLM
```

```python
# List of columns to convert
columns_to_convert = ['Gold_Price', 'Oil_Price', 'Nasdaq_Close', 'S&P500_Close', 'BTC_Close','BTC_Volume','BTC_Market_Cap','Tweets_Count','Tweets_Vader_Sentiment','News_Vader_Sentiment']

# Applying the transformation
for column in columns_to_convert:
    XLM[column + '_binary'] = (XLM[column] > XLM[column].shift(1)).astype(int)
```

```python
XLM.drop(["Internet_Penetration_USA","Population_USA","GDP_USA","Inflation_USA","Nasdaq_Open","Nasdaq_High","Nasdaq_Low","Nasdaq_Adj Close","Nasdaq_Volume","S&P500_Open","S&P500_High","S&P500_Low","S&P500_Adj Close","BTC_Open","BTC_High","BTC_Low",'Gold_Price', 'Oil_Price', 'Nasdaq_Close', 'S&P500_Close', 'BTC_Close','S&P500_Volume',],inplace=True,axis=1)
```

```python
XLM
```

```python
columns_to_lag = ['Stellar_Change'] #'Gold_Price_binary','Oil_Price_binary','Tweets_Count_binary','Tweets_Vader_Sentiment_binary','News_Vader_Sentiment_binary'

# Number of lags (days back) to create
number_of_lags = 5

# Create lag features for the selected columns
for column in columns_to_lag:
    for lag in range(1, number_of_lags + 1):
        XLM[f'{column}_lag_{lag}'] = XLM[column].shift(lag)

# Optionally, remove any rows with NaN values that were created due to shifting
XLM.dropna(inplace=True)
```

```python
XLM
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Selecting the columns to standardize
features = XLM[['BTC_Volume', 'BTC_Market_Cap', 'Tweets_Count', 'Tweets_Vader_Sentiment', 'News_Vader_Sentiment']]

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the features
features_scaled = scaler.fit_transform(features)

# Replace the original columns with the scaled columns
XLM[['BTC_Volume', 'BTC_Market_Cap', 'Tweets_Count', 'Tweets_Vader_Sentiment', 'News_Vader_Sentiment']] = features_scaled
```

```python
XLM
```

```python
XLM_market= pd.read_excel("../../Base_Datasets/Alt_Coin_Yearly_Rank.xlsx", sheet_name = 'Stellar_Daily')
XLM_market['Date'] = pd.to_datetime(XLM_market['Date'])
XLM_market.set_index('Date', inplace=True)
```

```python
XLM_market.tail(10)
```

```python
XLM_market
```

```python
# Convert columns to float, removing any non-numeric characters like commas
XLM_market['Open*'] = pd.to_numeric(XLM_market['Open*'].replace('[^\d.]', '', regex=True), errors='coerce')
XLM_market['High'] = pd.to_numeric(XLM_market['High'].replace('[^\d.]', '', regex=True), errors='coerce')
XLM_market['Low'] = pd.to_numeric(XLM_market['Low'].replace('[^\d.]', '', regex=True), errors='coerce')
XLM_market['Close**'] = pd.to_numeric(XLM_market['Close**'].replace('[^\d.]', '', regex=True), errors='coerce')
XLM_market['Volume'] = pd.to_numeric(XLM_market['Volume'].replace('[^\d.]', '', regex=True), errors='coerce')
XLM_market['Market Cap'] = pd.to_numeric(XLM_market['Market Cap'].replace('[^\d.]', '', regex=True), errors='coerce')
```

```python
XLM_market
```

```python
from ta import add_all_ta_features
from ta.utils import dropna

# Clean NaN values
XLM_market = dropna(XLM_market)

# Add all technical indicators
XLM_market = add_all_ta_features(
    XLM_market, open="Open*", high="High", low="Low", close="Close**", volume="Volume", fillna=True
)
```

```python
XLM_market
```

```python
from ta.momentum import RSIIndicator
from ta.trend import MACD
from ta.volatility import BollingerBands

# RSI
XLM_market['rsi'] = RSIIndicator(close=XLM_market['Close**'], window=14, fillna=True).rsi()

# MACD
macd = MACD(close=XLM_market['Close**'], window_slow=26, window_fast=12, window_sign=9, fillna=True)
XLM_market['macd'] = macd.macd()
XLM_market['macd_signal'] = macd.macd_signal()
XLM_market['macd_diff'] = macd.macd_diff()

# Bollinger Bands
indicator_bb = BollingerBands(close=XLM_market['Close**'], window=20, window_dev=2, fillna=True)
XLM_market['bb_bbm'] = indicator_bb.bollinger_mavg()
XLM_market['bb_bbh'] = indicator_bb.bollinger_hband()
XLM_market['bb_bbl'] = indicator_bb.bollinger_lband()
```

```python
XLM_market
```

```python
# Shift all columns except 'Date'
columns_to_shift = XLM_market.columns.difference(['Date'])
XLM_market[columns_to_shift] = XLM_market[columns_to_shift].shift(-1)

XLM_market.dropna(inplace=True)
```
