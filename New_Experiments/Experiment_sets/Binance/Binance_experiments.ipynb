```python
import pandas as pd
```

```python
BNB=pd.read_excel("../../../Binance/0 day lag individual.xlsx")
BNB['Date'] = pd.to_datetime(BNB['Date'])
BNB.set_index('Date', inplace=True)
```

```python
BNB
```

```python
#BTC MarketCap missing
# List of columns to convert
columns_to_convert = ['Gold_Price', 'Oil_Price','BTC_Volume','Tweets_Count','Tweets_Vader_Sentiment','News_Vader_Sentiment']

# Applying the transformation
for column in columns_to_convert:
    BNB[column + '_binary'] = (BNB[column] > BNB[column].shift(1)).astype(int)
```

```python
BNB
```

```python
BNB.drop(["Gold_Price","Oil_Price","Nasdaq_Volume","S&P500_Volume"],inplace=True,axis=1)
#'BTC_Volume','BTC_Market_Cap','Tweets_Count','Tweets_Vader_Sentiment','News_Vader_Sentiment'
```

```python
columns_to_lag = ['Binance_change'] #'Gold_Price_binary','Oil_Price_binary','Tweets_Count_binary','Tweets_Vader_Sentiment_binary','News_Vader_Sentiment_binary'

# Number of lags (days back) to create
number_of_lags = 5

# Create lag features for the selected columns
for column in columns_to_lag:
    for lag in range(1, number_of_lags + 1):
        BNB[f'{column}_lag_{lag}'] = BNB[column].shift(lag)

# Optionally, remove any rows with NaN values that were created due to shifting
BNB.dropna(inplace=True)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

#BTC market cap missing

# Selecting the columns to standardize
features = BNB[['BTC_Volume', 'Tweets_Count', 'Tweets_Vader_Sentiment', 'News_Vader_Sentiment']]

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the features
features_scaled = scaler.fit_transform(features)

# Replace the original columns with the scaled columns
BNB[['BTC_Volume', 'Tweets_Count', 'Tweets_Vader_Sentiment', 'News_Vader_Sentiment']] = features_scaled
```
