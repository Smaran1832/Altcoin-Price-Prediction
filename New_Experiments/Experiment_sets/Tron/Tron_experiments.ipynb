```python
import pandas as pd
```

```python
TRX=pd.read_excel("../../../Tron/0 day lag individual.xlsx")
TRX['Date'] = pd.to_datetime(TRX['Date'])
TRX.set_index('Date', inplace=True)
```

```python
TRX
```

```python
# btc market cap missing
# List of columns to convert
columns_to_convert = ['Gold_Price', 'Oil_Price','BTC_Volume','Tweets_Count','Tweets_Vader_Sentiment','News_Vader_Sentiment']

# Applying the transformation
for column in columns_to_convert:
    TRX[column + '_binary'] = (TRX[column] > TRX[column].shift(1)).astype(int)
```

```python
TRX.drop(["Nasdaq_Volume",'Gold_Price', 'Oil_Price','S&P500_Volume'],inplace=True,axis=1)
```

```python
TRX
```

```python
columns_to_lag = ['Tron_Change'] #'Gold_Price_binary','Oil_Price_binary','Tweets_Count_binary','Tweets_Vader_Sentiment_binary','News_Vader_Sentiment_binary'

# Number of lags (days back) to create
number_of_lags = 5

# Create lag features for the selected columns
for column in columns_to_lag:
    for lag in range(1, number_of_lags + 1):
        TRX[f'{column}_lag_{lag}'] = TRX[column].shift(lag)

# Optionally, remove any rows with NaN values that were created due to shifting
TRX.dropna(inplace=True)
```

```python
TRX
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Selecting the columns to standardize
features = TRX[['BTC_Volume', 'Tweets_Count', 'Tweets_Vader_Sentiment', 'News_Vader_Sentiment']]

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the features
features_scaled = scaler.fit_transform(features)

# Replace the original columns with the scaled columns
TRX[['BTC_Volume', 'Tweets_Count', 'Tweets_Vader_Sentiment', 'News_Vader_Sentiment']] = features_scaled
```

```python
TRX
```

```python
TRX_market= pd.read_excel("../../Base_Datasets/Alt_Coin_Yearly_Rank.xlsx", sheet_name = 'TRON_Daily')
TRX_market['Date'] = pd.to_datetime(TRX_market['Date'])
TRX_market.set_index('Date', inplace=True)
```

```python
TRX_market.tail(10)
```

```python
TRX_market
```

```python
# Convert columns to float, removing any non-numeric characters like commas
TRX_market['Open*'] = pd.to_numeric(TRX_market['Open*'].replace('[^\d.]', '', regex=True), errors='coerce')
TRX_market['High'] = pd.to_numeric(TRX_market['High'].replace('[^\d.]', '', regex=True), errors='coerce')
TRX_market['Low'] = pd.to_numeric(TRX_market['Low'].replace('[^\d.]', '', regex=True), errors='coerce')
TRX_market['Close**'] = pd.to_numeric(TRX_market['Close**'].replace('[^\d.]', '', regex=True), errors='coerce')
TRX_market['Volume'] = pd.to_numeric(TRX_market['Volume'].replace('[^\d.]', '', regex=True), errors='coerce')
TRX_market['Market Cap'] = pd.to_numeric(TRX_market['Market Cap'].replace('[^\d.]', '', regex=True), errors='coerce')
```

```python
TRX_market
```

```python
from ta import add_all_ta_features
from ta.utils import dropna

# Clean NaN values
TRX_market = dropna(TRX_market)

# Add all technical indicators
TRX_market = add_all_ta_features(
    TRX_market, open="Open*", high="High", low="Low", close="Close**", volume="Volume", fillna=True
)
```

```python
TRX_market
```

```python
from ta.momentum import RSIIndicator
from ta.trend import MACD
from ta.volatility import BollingerBands

# RSI
TRX_market['rsi'] = RSIIndicator(close=TRX_market['Close**'], window=14, fillna=True).rsi()

# MACD
macd = MACD(close=TRX_market['Close**'], window_slow=26, window_fast=12, window_sign=9, fillna=True)
TRX_market['macd'] = macd.macd()
TRX_market['macd_signal'] = macd.macd_signal()
TRX_market['macd_diff'] = macd.macd_diff()

# Bollinger Bands
indicator_bb = BollingerBands(close=TRX_market['Close**'], window=20, window_dev=2, fillna=True)
TRX_market['bb_bbm'] = indicator_bb.bollinger_mavg()
TRX_market['bb_bbh'] = indicator_bb.bollinger_hband()
TRX_market['bb_bbl'] = indicator_bb.bollinger_lband()
```

```python
TRX_market
```

```python
# Shift all columns except 'Date'
columns_to_shift = TRX_market.columns.difference(['Date'])
TRX_market[columns_to_shift] = TRX_market[columns_to_shift].shift(-1)

TRX_market.dropna(inplace=True)
```

```python
Tron_final_dataset=TRX_market.join(TRX,how='inner') 
```
